# Connect with MySQL database
### Read Clean_Data.csv file 

```python
import pandas as pd
df = pd.read_csv("Clean_Data.csv",index_col = [0])
```

### Rename the columns name in dataframe

```python
df.rename(columns={"CMP Rs.":"Current_Price","P/E":"P2E_Ratio","Div Yld %":"Div_Yld","ROCE %":"ROCE"},inplace=True)
df
```

### Import required Libraries 

```python
import sqlalchemy
import pymysql
```

### Create SQLAlchemy engine to connect to MySQL Database

```python
# engine = create_engine("mysql+pymysql://{user}:{pw}@{host}/{db}".format(host=hostname, db=dbname, user=uname, pw=pwd))

engine = sqlalchemy.create_engine("mysql+pymysql://root:@localhost:3306/fundamental_stocks")
```

### Export dataframe data to MySQL database as form of table

```python
# Convert dataframe to sql table

df.to_sql('stocks', engine, index=False,if_exists="replace") # fail, replace, append
```

```python
query = '''SELECT * FROM fundamental_stocks.stocks where ROCE>20 and Div_Yld>2.75 order by ROCE desc limit 30 ;'''
```

### Execute MySQL query 
With pd.read_sql() function, we execute the 'select' sql query on table 'stocks' to get desired output.

```python
pdf = pd.read_sql(query,engine)
pdf
```

### Export Final Resultant data from "pdf" dataframe to a "Final_Data.csv" file.

```python
pdf.to_csv("Final_Result.csv",index=False)
```
